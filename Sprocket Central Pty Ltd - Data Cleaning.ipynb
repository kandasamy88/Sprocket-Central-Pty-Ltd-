we will create features engineering for some columns

Changing DOB column into Age column

Dropping DOB after conversion

```python
# This function converts given date to age
def from_dob_to_age(born):
    today = dt.date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
#applying the function on the DOB column
Custdem['Age']=Custdem['DOB'].apply(lambda x: from_dob_to_age(x))
```
